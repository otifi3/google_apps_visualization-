```r
ggplot(google_apps_data)+
  geom_bar(aes(x = as.factor(months),fill = Type), position = position_stack(reverse = TRUE))+
  theme(legend.position = "right")+
  theme_linedraw()+
  labs(title = "Type App Added Over The Month",
         x = "Months",
         y = "Count")
```

```r
ggplot(google_apps_data)+
  geom_bar(aes(x = forcats::fct_infreq(Category),fill = Type),
           position = position_stack(reverse = T))+
  coord_flip()+
  theme(legend.position = "right")+
  theme_linedraw()+
  labs(title = "Category Android App With Type", 
       y="Frequency",
       x="Category")
```

```r
ggplot(google_apps_data,aes(y= Rating, x=Price)) + 
  geom_line(size = 1,color = "orange")+
  theme_linedraw()+
  labs(title = "Effect of Price on App Rating",
         x = "Price",
         y = "Rating")
```

```r
ggplot(data = google_apps_data, aes (x=Content_Rating, y = Rating, fill = Content_Rating))+
  geom_boxplot(size = 0.5, outlier.color = "red", outlier.size = 1)+
  stat_boxplot(geom = "errorbar", size = 0.5)+
  stat_summary(fun.y = mean, col = "black", geom = "point")+
  theme(legend.position = "none")+
  labs(title = "Content Rating vs Rating", 
         x = "Content Rating",
         y = "Rating")
```

```r
ggplot(google_apps_data, aes(x=Reviews, y=Rating)) +
  scale_x_continuous(trans='log10', labels=comma) +
  geom_point(aes(col=Type)) +
  stat_smooth(method = "lm", col = "blue", size = 2)+
  labs(title="Android App Ratings vs Number of Reviews", y="Rating", x="Number of Reviews")+
  theme_linedraw()
```

```r
google_apps_data  %>%
select(Category,Rating) %>% group_by(Category) %>%
mutate(median=median(Rating,na.rm=TRUE)) %>% 
ggplot(aes(x=fct_reorder(Category,median),y=Rating)) + geom_boxplot(alpha=0.8,fill="#69b3a2") + coord_flip() +  labs(title="Category vs Rating",y="Rating",x="Category")
```
