```r
names(google_apps_data)[9] <- "Content_Rating"
names(google_apps_data)[11] <- "Last_Updated"
names(google_apps_data)[12] <- "Current_Version"
names(google_apps_data)[13] <- "Andriod_Version"
```

```r
google_apps_data <- google_apps_data %>% mutate(Price = as.factor(gsub("\\$", "", Price)))
```

```r
google_apps_data$Price <- as.numeric(as.character(google_apps_data$Price))
```

```r
google_apps_data <- google_apps_data %>% mutate(Size = as.numeric(gsub("M", "", Size)))
```

```r
google_apps_data <- google_apps_data %>% mutate(Reviews = as.numeric(as.character(Reviews)))
```

```r
google_apps_data <- google_apps_data %>% mutate(Rating = as.factor(as.numeric(Rating)))
```

```r
google_apps_data <- google_apps_data %>% mutate(Installs = as.factor(gsub(",", "", Installs)))
```

```r
google_apps_data <- google_apps_data %>% mutate(Installs = as.factor(gsub("\\+", "", Installs)))
```

```r
google_apps_data <- google_apps_data %>% separate(Last_Updated, c("Days_months", "Years"), ",")
```

```r
str(google_apps_data)
```

```r
head(google_apps_data)
```

```r
google_apps_data <- google_apps_data %>% separate(Days_months, c("Months", "Days"), " ")
```

```r
google_apps_data$Months <- match(google_apps_data$Months, month.name)
```

```r
google_apps_data$Date <- as.Date(with(google_apps_data,paste(Days,Months,Years,sep="-")),"%d-%m-%Y")
```
