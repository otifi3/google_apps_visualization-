```python
fig = px.histogram(data_frame=df.sort_values(by='Rating'), x='Category', y="Reviews",histfunc="sum",
                   color_discrete_sequence=['cyan'])

fig.update_layout(font=dict(family='Lato', size=18, color='black'), 
                  title=dict(text='Relation between category and number of reviews',
                           font=dict(size=30), x=.5), 
                  paper_bgcolor= 'white', plot_bgcolor='white',
                 xaxis = dict(title='Categorys', showgrid=False),
                 yaxis=dict(showgrid=False),bargap=0.2)   
```

```python
df2.dropna(axis=0, subset=["Translated_Review"], inplace=True)
```

```python
col = "Sentiment"
grouped = df2[col].value_counts().reset_index()
grouped = grouped.rename(columns = {col : "count", "index" : col})

trace = go.Pie(labels=grouped[col], values=grouped['count'], pull=[0.05, 0], marker=dict(colors=["#6ad49b", "#a678de"]))
layout = go.Layout(title="", height=400, legend=dict(x=0.1, y=1.1))
fig = go.Figure(data = [trace], layout = layout)
iplot(fig)
```

```python
df = px.data.gapminder().query("year==2018")
fig = px.treemap(df, path=[px.Constant("Content"), 'Content Rating', 'Category'], values='Installs',
                  color='Reviews', hover_data=['iso_alpha'],
                  color_continuous_scale='RdBu',
                  color_continuous_midpoint=np.average(df['Reviews'], weights=df['Installs']))
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```
