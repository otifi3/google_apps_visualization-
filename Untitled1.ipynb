```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
apps = pd.read_csv("Data.csv")
```

```python
Sentiment = pd.read_csv("Sentiment.csv")
```

```python
fig = px.treemap(apps, path=[ 'Category', 'Android_Version'], values='Installs', color='Rating', color_continuous_scale='Inferno')
fig.update_traces(root_color="White")
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
fig = px.line(apps, x="months", y="Installs", color='Years',linefunc = "sum")
fig.show()
```

```python
fig = px.line_polar(Sentiment, r=[0,0.25,0.5,0.75,1], theta=['Sentiment_Subjectivity_Neutral','Sentiment_Subjectivity_Negative','Sentiment_Subjectivity_Positive','Sentiment_Polarity_Positive','Sentiment_Polarity_Negative'],bline_close=True)
fig.update_traces(fill='toself')
fig.show()
```

```python
fig = px.histogram(data_frame=apps.sort_values(by='Rating'), x='Category', y="Reviews",histfunc="sum",
                   color_discrete_sequence=['Black'])

fig.update_layout(font=dict(family='Lato', size=18, color='black'), 
                  title=dict(text='Relation between category and number of reviews',
                           font=dict(size=30), x=.5), 
                  paper_bgcolor= 'white', plot_bgcolor='white',
                 xaxis = dict(title='Categories', showgrid=False),
                 yaxis=dict(showgrid=False),bargap=0.2)   
```

```python


apps["normalized_installs"] = (apps["Installs"]-apps["Installs"].mean())/apps["Installs"].std()
px.scatter(data_frame = apps
           ,x = 'normalized_installs'
           ,y = 'Rating'
           ,color = 'Type'
           )
```
